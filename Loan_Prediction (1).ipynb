# Dealing with Outliers

```python
import numpy as np
q1 = df['ApplicantIncome'].quantile(0.25)
q3 = df['ApplicantIncome'].quantile(0.75)
iqr = q3 - q1
low = q1 - 1.5*iqr
high = q3 + 1.5*iqr


df['ApplicantIncome'] = np.where(df['ApplicantIncome'] < low, low, df['ApplicantIncome'])
df['ApplicantIncome'] = np.where(df['ApplicantIncome'] > high, high, df['ApplicantIncome'])
```

```python
q1 = df['CoapplicantIncome'].quantile(0.25)
q3 = df['CoapplicantIncome'].quantile(0.75)
iqr = q3 - q1
low = q1 - 1.5*iqr
high = q3 + 1.5*iqr


df['CoapplicantIncome'] = np.where(df['CoapplicantIncome'] < low, low, df['CoapplicantIncome'])
df['CoapplicantIncome'] = np.where(df['CoapplicantIncome'] > high, high, df['CoapplicantIncome'])
```

# spliting data into train and test  and Validate

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
x1, x2, y1, y2 = train_test_split(X,Y,test_size = 0.2,random_state = 42)
x1, x3, y1, y3 = train_test_split(x1,y1, test_size=0.25, random_state=42)
```

# Model Fitting

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr
```

```python
lr.fit(x1,y1)
y_pred = lr.predict(x2)
y_pred
```

```python
lr.score(x2,y2)*100
```

#   # Evaluate the model on the training, validation, and testing sets

```python
from sklearn.metrics import mean_squared_error

train_error = mean_squared_error(y1, lr.predict(x1))
val_error = mean_squared_error(y3, lr.predict(x3))
test_error = mean_squared_error(y2, lr.predict(x2))
```
